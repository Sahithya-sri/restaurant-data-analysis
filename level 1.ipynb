```python
# Determine the top three most common cuisines in the dataset.

# Import necessary libraries
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv(r'C:\Users\Srinivasan\Downloads/Dataset .csv')

# Group by 'Cuisine' column and count occurrences
cuisine_counts = df.groupby('Cuisines').size().reset_index(name='Count')

# Sort in descending order by 'Count'
cuisine_counts = cuisine_counts.sort_values(by='Count', ascending=False)

# Select the top 3 rows
top_3_cuisines = cuisine_counts.head(3)

# Print the top 3 most common cuisines
print(top_3_cuisines)

```

```python
Task 2 : City Analysis
```

```python
Q : Calculate the percentage of restaurants that serve each of the top cuisines.
```

```python
# Import necessary libraries
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv(r'C:\Users\Srinivasan\Downloads/Dataset .csv')

# Get the total number of restaurants
total_restaurants = len(df)

# Get the top 3 most common cuisines
cuisine_counts = df.groupby('Cuisines').size().reset_index(name='Count')
cuisine_counts = cuisine_counts.sort_values(by='Count', ascending=False)
top_3_cuisines = cuisine_counts.head(3)

# Calculate the percentage for each of the top 3 cuisines
for index, row in top_3_cuisines.iterrows():
    cuisine = row['Cuisines']
    count = row['Count']
    percentage = (count / total_restaurants) * 100
    print(f"{cuisine} cuisine: {percentage:.2f}% of restaurants")
```

```python
Q : Identify the city with the highest number of restaurants in the dataset.
```

```python
# Import necessary libraries
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv(r'C:\Users\Srinivasan\Downloads/Dataset .csv')

# Group by 'City' column and count occurrences
city_counts = df.groupby('City').size().reset_index(name='Count')

# Sort in descending order by 'Count'
city_counts = city_counts.sort_values(by='Count', ascending=False)

# Get the city with the highest count
city_with_max_restaurants = city_counts.iloc[0]['City']
max_restaurant_count = city_counts.iloc[0]['Count']

# Print the result
print(f"The city with the highest number of restaurants is {city_with_max_restaurants} with {max_restaurant_count} restaurants.")
```

```python
import pandas as pd

# Load the dataset (assuming it's in a CSV file)
dataset = pd.read_csv(r'C:\Users\Srinivasan\Downloads/Dataset .csv')

# Display the first few rows of the dataset to understand its structure
print(dataset.head())

# Assuming the dataset has columns named 'city' and 'rating'
# Group by city and calculate the average rating
city_avg_ratings = dataset.groupby('City')['Rating text'].mean()

# Find the city with the highest average rating
city_with_highest_avg_rating = city_avg_ratings.idxmax()
highest_avg_rating = city_avg_ratings.max()

print(f"The city with the highest average rating is {city_with_highest_avg_rating} with an average rating of {highest_avg_rating}.")
```

```python
Q : Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv(r'C:\Users\Srinivasan\Downloads/Dataset .csv')

# Extract the 'Price Range' column and count occurrences
price_range_counts = df['Price range'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(price_range_counts.index, price_range_counts.values)
plt.xlabel('Price range')
plt.ylabel('Count')
plt.title('Distribution of Price ranges')
plt.xticks(rotation=90)
plt.show()
```

```python
Q : Calculate the percentage of restaurants in each price range category.
```

```python
# Import necessary libraries
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv(r'C:\Users\Srinivasan\Downloads/Dataset .csv')

# Count the occurrences of each price range category
price_range_counts = df['Price range'].value_counts()

# Calculate the total number of restaurants
total_restaurants = len(df)

# Calculate the percentage for each price range category
price_range_percentages = (price_range_counts / total_restaurants) * 100

# Print the result
print("Percentage of restaurants in each price range category:")
for price_range, percentage in price_range_percentages.items():
    print(f"{price_range}: {percentage:.2f}%")
```
